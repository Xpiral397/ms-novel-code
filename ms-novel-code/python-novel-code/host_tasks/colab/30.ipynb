# Prompt

You are given a plank of length n, where the positions on the plank range from 0 to n. A number of ants are placed at integer positions on the plank. Each ant moves at a constant speed of 1 unit per second, either left or right.
- When two ants meet, they instantly reverse directions.
- However, for the purpose of calculating fall-off times, this is equivalent to ants simply passing through each other without changing direction.
- An ant falls off the plank immediately upon reaching position 0 or position n.

You are given:
- An integer n, representing the length of the plank.
- A list of positions where ants are initially moving to the left.
- A list of positions where ants are initially moving to the right.

Return the time in seconds at which the last ant falls off the plank.

Constraints:

- 1 <= n <= 10^6
- 0 <= l1, l2 <= 100000
- 0 <= l1[i], l2[j] <= n
- All positions are distinct
- Ants that start at position 0 or n fall off immediately

Example 1:

Input:
n = 10
l1 = [2, 6]
l2 = [7]

Output:
6

Explanation:
- Ant at position 2 moving left falls off at time 2
- Ant at position 6 moving left falls off at time 6
- Ant at position 7 moving right falls off at time 3
The last fall is at time 6

Example 2:

Input:
n = 100
l1 = [10, 50, 90]
l2 = [0, 100]

Output:
90

Explanation:
- Right-moving ants at positions 0 and 100 fall off immediately
- Left-moving ants at 10, 50, 90 fall off at times 10, 50, 90
Last fall is at 90

# Requirements

## Explicit Requirements:
- You are given an integer n which represents the length of the plank.
- You are given a list of positions of ants moving to the left.
- You are given a list of positions of ants moving to the right.
- All ants move at a speed of 1 unit per second.
- Ants fall off the plank immediately upon reaching position 0 or position n.
- You must return the time in seconds when the last ant falls off the plank.

## Implicit Requirements:
- When ants meet, they reverse direction, but for the purpose of this problem, this can be treated as them passing through each other.
- The fall-off time for a left-moving ant is equal to its position.
- The fall-off time for a right-moving ant is equal to n minus its position.
- The result is the maximum fall-off time among all ants.
- Ants at position 0 or n fall off immediately at time 0.

## Edge Cases:
- If both the left and right lists are empty, return 0.
- If all ants are already at position 0 or n, return 0.
- If one of the lists is empty, calculate the fall-off time using only the
non-empty list.

## Function Signature
```python
def last_ant_fall_time(n: int, l1: List[int], l2: List[int]) -> int:
```

## Constraints:
- 1 <= n <= 10^6
- 0 <= l1, l2 <= 100000
- 0 <= l1[i], l2[j] <= n
- All positions are distinct
- Ants that start at position 0 or n fall off immediately

````python
# code
from typing import List
def last_ant_fall_time(n: int, l1: List[int], l2: List[int]) -> int:
    if not l1 and not l2:
      return 0

    l2 = [n-i for i in l2]

    if not l2:
        return max(l1)
    if not l1:
        return max(l2)
    answer = max(max([i for i in l1 if i not in [0, n]]), max([i for i in l1 if i not in [0, n]]))
    return answer

# Examples
print(last_ant_fall_time(10, [2, 6], [7]))  # Output: 6
print(last_ant_fall_time(100, [10, 50, 90], [0, 100]))  # Output: 90
````
